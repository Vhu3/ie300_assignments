### Five cards are drawn from a standard 52 card deck of playing cards,  each card is equally likely.
<img src='std52.png'>


<ol>
    <li> <b> TWO PAIR  </b> is the event that:
        <ul>
            <li> two cards both have one number, </li>
            <li> two other cards both have some other number, </li>
            <li> the fifth card has a number different from the other two numbers. </li>
        </ul>
            Find $P$(TWO PAIR). </li>
    <li> <b> THREE OF A KIND </b> is the event that:
        <ul>
            <li> three of the cards all have the same number, </li>
            <li> the remaining two cards have numbers different from each other and the three with
                the same number. </li>
        </ul> 
        Find $P$(THREE OF A KIND). </li>
    
<li> <b> FOUR OF A KIND </b> is the event that four of the five cards have the same number. 
    Find $P$(FOUR OF A KIND). </li>
<ol>
    


# Solution


<ol>
    <li> <b> TWO PAIR:  </b>
        <ul>
            <li> Number of ways to choose 5 cards from 52 possible cards: $52 \choose 5$ </li>
            <li> Number of ways to choose 2 distinct numbers for the two pairs from 13 possible numbers: $13 \choose 2$ </li>
            <li> Number of ways to choose 2 distinct suits from 4 possible suits for the first pair: $4 \choose 2$ </li>
            <li> Number of ways to choose 2 distinct suits from 4 possible suits for the second pair: $4 \choose 2$ </li>
            <li> Number of ways to choose 1 number from the 11 remaining numbers for the fifth card: ${11 \choose 1} = 11$ </li>
            <li> Number of ways to choose 1 suit from 4 possible suits for the fifth card: ${4 \choose 1} = 4$ </li>
        </ul>
            Then $P$(TWO PAIR)$=\dfrac{{13 \choose 2}{4 \choose 2}{4 \choose 2}\cdot 11 \cdot 4}{52 \choose 5}\approx 0.0475$. </li>
        <li> <b> THREE OF A KIND:  </b>
        <ul>
            <li> Number of ways to choose 5 cards from 52 possible cards: $52 \choose 5$ </li>
            <li> Number of ways to choose 1 number for the three of a kind from 13 possible numbers: ${13 \choose 1}=13$ </li>
            <li> Number of ways to choose 3 distinct suits from 4 possible suits for the three of a kind: ${4 \choose 3}=4$ </li>
            <li> Number of ways to choose 2 distinct numbers from the 12 remaining numbers for the fourth and fifth cards: ${12 \choose 2}$ </li>
            <li> Number of ways to choose 1 suit from 4 possible suits for the fourth card: ${4 \choose 1} = 4$ </li>
            <li> Number of ways to choose 1 suit from 4 possible suits for the fifth card: ${4 \choose 1} = 4$ </li>
        </ul>
            Then $P$(THREE OF A KIND)$=\dfrac{13\cdot 4\cdot {12 \choose 2}\cdot 4\cdot 4 }{52 \choose 5}=\dfrac{13\cdot {12 \choose 2}\cdot 4^3 }{52 \choose 5}\approx 0.0211$. </li>
    <li> <b> FOUR OF A KIND:  </b>
        <ul>
            <li> Number of ways to choose 5 cards from 52 possible cards: $52 \choose 5$ </li>
            <li> Number of ways to choose 1 number for the four of a kind from 13 possible numbers: ${13 \choose 1}=13$ </li>
            <li> Number of ways to choose 1 number from the 12 remaining numbers for the fifth card: ${12 \choose 1}=12$ </li>
            <li> Number of ways to choose 1 suit from 4 possible suits for the fifth card: ${4 \choose 1} = 4$ </li>
        </ul>
            Then $P$(FOUR OF A KIND)$=\dfrac{13\cdot 12\cdot 4 }{52 \choose 5}\approx 0.0002401$. </li>

```python
# Part 1: generate cards
suits = ['H','D','S','C']  # Hearts, diamonds, spades, clubs ♥♦♠♣
values = [str(n) for n in range(2,11)]+['J','Q','K','A']
cards = [v+s for s in suits for v in values]
print(cards)
```

```python
# This is a technical part for fancy graphics -  no need to explain it
from matplotlib.pyplot import subplots
import numpy as np
from PIL import Image
%matplotlib inline
print('Loading icons')
pil_im = {s:np.asarray(Image.open(f'cards/%s.png'%s, 'r')) for s in cards}
print('done')

def show_cards(cds):
    fig,ax = subplots(1,len(cds),figsize=(5*len(cds),5))
    for n,c in enumerate(cds):
        ax[n].imshow(pil_im[c])
        ax[n].axis('off')       
```

```python
# Part 2
def subset2(L):
    sub2 = []
                             
    for n1, v1 in enumerate(L):
        tail1 = L[n1+1:]
        for n2, v2 in enumerate(tail1):
            sub2.append({v1,v2})

# Alternative implementation:
#
#    for i in range(0,len(L)):
#        for j in range(0,len(L)):
#            if j > i:
#                sub2.append({L[i],L[j]})
            
            
    return sub2

print(subset2([1,2,3,4]))
```

```python
# Part 3

from itertools import combinations

# combinations(X,k) creates a list (iterable) of all subset of size k of X
# Essentially like our subset2 from the previous part, only more general, and better

two_pair = []
tot = 0
for c1,c2,c3 in combinations(values,3):
    for s11,s12 in combinations(suits,2):
        for s21,s22 in combinations(suits,2):
            for s3 in suits:
                two_pair.append((c1+s11, c1+s12, c2+s21, c2+s22, c3+s3))
                two_pair.append((c2+s11, c2+s12, c3+s21, c3+s22, c1+s3))
                two_pair.append((c3+s11, c3+s12, c1+s21, c1+s22, c2+s3))
#                two_pair.append((c1+s21, c1+s22, c2+s11, c2+s12, c3+s3))

# Show a couple of generated two-pairs
                
show_cards(two_pair[4321])
show_cards(two_pair[1234])

print(len(two_pair))

# Let's compare with our result from the theoretical part:
from math import factorial 
def binom(m,n):
    return factorial(m)//(factorial(n)*factorial(m-n))

print(binom(13,2)*11*36*4)
```

```python
# Part 4

three_of_a_kind = []
for c1,c2,c3 in combinations(values,3):
    for s11,s12,s13 in combinations(suits,3):
        for s2 in suits:
            for s3 in suits:
                three_of_a_kind.append((c1+s11, c1+s12, c1+s13, c2+s2, c3+s3))
                three_of_a_kind.append((c2+s11, c2+s12, c2+s13, c3+s2, c1+s3))
                three_of_a_kind.append((c3+s11, c3+s12, c3+s13, c1+s2, c2+s3))

# Show a couple of generated two-pairs
                
show_cards(three_of_a_kind[4321])
show_cards(three_of_a_kind[1234])


print(len(three_of_a_kind))

print(binom(12,2)*13*4*4*4)
```

```python
# Part 5

four_of_a_kind = []
for c1,c2 in combinations(values,2):
    for s11,s12,s13,s14 in combinations(suits,4):
        for s2 in suits:
            four_of_a_kind.append((c1+s11, c1+s12, c1+s13, c1+s14, c2+s2))
            four_of_a_kind.append((c2+s11, c2+s12, c2+s13, c2+s14, c1+s2))

# Show a couple of generated two-pairs
                
show_cards(four_of_a_kind[123])
show_cards(four_of_a_kind[321])


print(len(four_of_a_kind))

print(13*12*4)
```
